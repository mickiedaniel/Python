```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
##le o conteudo da url
url = 'https://en.wikipedia.org/wiki/Pulitzer_Prize_for_Drama'
response = requests.get(url)
html = response.content
```

```python
##faz o parse da pagina para printar de forma legivel
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
##os titulos dos ganhadores estavam em bold então pego todos os 'b'
bold = soup.find_all('b')
print(bold)
```

```python
##crio um arquivo .txt e escrevo conforme o for
##se estiver vazio ele n escreve
with open('saida.txt', 'w') as arquivo:
    for row in soup.find_all('b'):
        a_tag = row.find('a')
        if a_tag is not None and a_tag.get('title') is not None:
            arquivo.write(a_tag.get('title') + '\n')
```

```python
with open('saida.txt', 'r') as arquivo:
    linhas = arquivo.readlines()
    for linha in linhas:
        print(linha)
```
